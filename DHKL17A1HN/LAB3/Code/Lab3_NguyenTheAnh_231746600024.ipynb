LAB 3.2. XỬ LÝ DỮ LIỆU NULL (thời gian thực hiện: 15-20 phút.)
Mục tiêu: Học cách xác định và xử lý dữ liệu Null.
Nhiệm Vụ:
1. Kiểm tra xem trong stocks1 có dữ liệu Null nào không.
2. Thay thế dữ liệu Null ở cột high bằng giá trị trung bình của cột high.
3. Thay thế dữ liệu Null ở cột low bằng giá trị trung bình của cột low.
4. Hiển thị thông tin tổng quan để xác nhận không còn dữ liệu Null.

```python

#Kiểm tra dữ liệu Null trong stocks1
Ktra_Null= stocks1.isnull()
print(Ktra_Null.sum()) 

#Tính gtri trung bình và thay thế gtri Null
gtri_tb=stocks1['high'].mean()
stocks1['high'] = stocks1['high'].fillna(gtri_tb)
print(stocks1['high'])


#Tính gtri trung bình và thay thế gtri Null
gtri_tbinh=stocks1['low'].mean()
stocks1['low'] = stocks1['low'].fillna(gtri_tbinh)
print(stocks1['low'])

#Hiển thị tổng quan sau khi thay thế gtri
stocks1.info()
```

LAB 3.3. GỘP DỮ LIỆU VÀ PHÂN TÍCH CƠ BẢN (thời gian thực hiện: 15-20 phút.)
Mục tiêu: Nắm vững cách gộp dữ liệu và thực hiện phân tích cơ bản.
Nhiệm Vụ:
1. Đọc file stocks2.csv vào DataFrame stocks2.
2. Gộp stocks1 và stocks2 thành DataFrame mới tên là stocks.
3. Tính giá trung bình (open, high, low, close) cho mỗi ngày.
4. Hiển thị 5 dòng đầu tiên của kết quả.


```python
file='H:\DHKL17A1HN\LAB3\Data\stocks2.csv'
stocks2=pd.read_csv(file)

#Gộp stocks1 và stocks2
stocks=pd.concat([stocks1,stocks2],ignore_index=True)
print(stocks)

#Tính gtri trung bình cho mỗi ngày
stocks['average'] = stocks[['open', 'high', 'low', 'close']].mean(axis=1)
print(stocks)
stocks[:5]
```
